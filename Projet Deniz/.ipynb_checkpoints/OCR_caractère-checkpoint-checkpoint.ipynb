# Traitement d'image de caract√®re

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage import filters, img_as_ubyte
from sklearn.decomposition import PCA
import os
from scipy.spatial import distance
from mpl_toolkits.mplot3d import Axes3D



### Fonctions de traitement d'image ###

def mat_gris(img_path): 
    # convertit l'image en matrice de niveaux de gris
    # chaque pixel est associ√© √† une valeur allant de 0 (noir) √† 255 (blanc)
    img = Image.open(img_path)
    img_gris = img.convert('L')
    mat_gris = np.array(img_gris)
    return mat_gris

def histogramme(mat):
    # cr√©e et trace un histogramme de niveaux de gris de l'image
    mat1D = mat.flatten()
    plt.hist(mat1D, bins=256, color='gray', alpha=0.7)
    plt.show()

def seuillage(mat):
    # choisit un seuil via le m√©thode d'Otsu
    # chaque pixel est compar√© au seuil 
    # ceux de niveaux de gris inf√©rieurs sont mis √† 0 (noir), ceux sup√©rieurs mis √† 255 (blanc)
    seuil = filters.threshold_otsu(mat)
    for i in range(len(mat)):
        for j in range(len(mat[0])):
            if mat[i][j] < seuil:
                mat[i][j] = 0
            else:
                mat[i][j] = 255
    return mat

def contourer(mat):
    # contoure le caract√®re en utilisant le filtre de Sobel
    g = filters.sobel(mat)
    g = (g / np.max(g)) * 2 - 1 
    g = img_as_ubyte(g)
    g = 255 - g
    g = seuillage(g)
    return g

def rogner(mat):
    # rogne au maximum l'image en cherchant les limites de l'image
    mat = seuillage(mat)
    mat = contourer(mat)
    c_non_vides = np.where(np.any(mat==0, axis=0))[0]
    l_non_vides = np.where(np.any(mat==0, axis=1))[0]
    c_min, c_max = c_non_vides[0], c_non_vides[-1]
    l_min, l_max = l_non_vides[0], l_non_vides[-1]
    mat_rognee = mat[l_min:l_max+1, c_min:c_max+1]
    return mat_rognee

def reduire(mat, taille):
    # r√©duit la taille de l'image √† la taille donn√©e
    img = Image.fromarray(mat)
    img_reduite = img.resize((taille, taille))
    mat_reduite = np.array(img_reduite)
    return seuillage(mat_reduite)



### Affichage des traitements d'image ###

img_path = "scan_niv_gris_400ppp_fiche_ocr_P.bmp"
original_img = Image.open(img_path)
plt.figure(figsize=(6, 6))
plt.imshow(original_img, cmap='gray')
plt.title('Image originale')
plt.show()

gris_img = mat_gris(img_path)
rognee_img = rogner(gris_img)
reduit_img = reduire(rognee_img, 30)

plt.imshow(rognee_img, cmap='gray')
plt.title('Image seuill√©, contour√©e et rogn√©e')
plt.show()

plt.figure(figsize=(6, 6))
plt.imshow(reduit_img, cmap='gray')
plt.title(f'Image r√©duite')
plt.show()



### Projection 2D de l'alphabet et du caract√®re ###

def project_to_2d(mat):
    # Projette dans un espace 2D la matrice de chaque lettre
    pca = PCA(n_components=2)
    if len(mat.shape) > 2:
        mat = mat.reshape(mat.shape[0], -1)
    mat_2d = pca.fit_transform(mat)
    return mat_2d

lettres = [chr(i) for i in range(65, 91)] + [img_path]
matrices = []

# Parcours toutes les lettres de l'alphabet
for lettre in lettres:
    img_path = f"{lettre}.png"
    if os.path.isfile(img_path):
        mat = mat_gris(img_path)
        mat = rogner(mat)
        mat = reduire(mat, 30)
        matrices.append(mat.flatten())

matrices = np.array(matrices)
pca = PCA(n_components=2)
matrices_2d = pca.fit_transform(matrices)

plt.figure(figsize=(8, 6))
plt.scatter(matrices_2d[:, 0], matrices_2d[:, 1])

for i, lettre in enumerate(lettres):
    if i < len(matrices_2d):
        plt.annotate(lettre, (matrices_2d[i, 0], matrices_2d[i, 1]))

# Ajouter la matrice de l'image √† l'ensemble de donn√©es

mat = mat_gris(img_path)
mat = rogner(mat)
mat = reduire(mat, 30)
matrices = np.append(matrices, mat.flatten().reshape(1, -1), axis=0)

pca = PCA(n_components=2)
matrices_2d = pca.fit_transform(matrices)

mat_2d = matrices_2d[-1]

plt.scatter(mat_2d[0], mat_2d[1], color='red')
plt.annotate('test', (mat_2d[0], mat_2d[1]))

distances = distance.cdist(mat_2d.reshape(1, -1), matrices_2d[:-1]) 
closest_point_index = np.argmin(distances)

closest_letter = lettres[closest_point_index]

print(f"La lettre la plus proche est : {closest_letter}")

plt.title('Projection 2D')
plt.show()



### Projection 3D de l'alphabet et du caract√®re ###

def project_to_3d(mat):
    # Projette dans un espace 3D la matrice de chaque lettre
    pca = PCA(n_components=3)
    if len(mat.shape) > 2:
        mat = mat.reshape(mat.shape[0], -1)
    mat_3d = pca.fit_transform(mat)
    return mat_3d

matrices_3d = project_to_3d(matrices)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(matrices_3d[:-1, 0], matrices_3d[:-1, 1], matrices_3d[:-1, 2])

for i, lettre in enumerate(lettres):
    if i < len(matrices_3d) - 1:
        ax.text(matrices_3d[i, 0], matrices_3d[i, 1], matrices_3d[i, 2], lettre)

mat_3d = matrices_3d[-1]

ax.scatter(mat_3d[0], mat_3d[1], mat_3d[2], color='red')
ax.text(mat_3d[0], mat_3d[1], mat_3d[2], 'test')

distances = distance.cdist(mat_3d.reshape(1, -1), matrices_3d[:-1]) 
closest_point_index = np.argmin(distances)

closest_letter = lettres[closest_point_index]

print(f"La lettre la plus proche est : {closest_letter}")

plt.title('Projection 3D')
plt.show()
```
