```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from separation_en_caract√®res import separe_en_caracteres
from separation_en_lignes import separe_en_lignes
from skimage.io import imsave
```

```python
image = cv2.imread("scan_niv_gris_300ppp_fiche_ocr.bmp")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
(thresh, image_binary) = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
indices_lignes = separe_en_lignes(image_binary, reduction=2)
print(len(indices_lignes))
```

```python
for i in range(len(indices_lignes)) :

    indices_debut_fin_ligne = indices_lignes[i]

    output_dir = f'ligne{i}'
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    ranges = separe_en_caracteres(image_binary, indices_debut_fin_ligne)
    for elt in range(len(ranges)) :
        #plt.figure(figsize=(20,10))
        imsave(f'ligne{i}/char_{elt}.png', image[indices_debut_fin_ligne[0]:indices_debut_fin_ligne[1], ranges[elt][0]:ranges[elt][1]])
        #plt.imshow(image[indices_debut_fin_ligne[0]:indices_debut_fin_ligne[1], ranges[elt][0]:ranges[elt][1]])
        #plt.show()
        #plt.close()
```
